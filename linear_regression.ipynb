# **Features selections**

```python
features = ['trip_distance', 'fare_amount', 'extra', 'mta_tax', 'tip_amount',
       'tolls_amount', 'improvement_surcharge', 'duration']
```

## **Splitting data into X and Y**

```python
X = data_frame[features]
Y = data_frame['total_amount']
```

```python
X.head()
```

```python
Y.head()
```

**X, Y -> X_train, Y_train, X_valid, Y_valid**

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
data_frame.shape
```

# **Training linear regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regression = LinearRegression()

linear_regression.fit(X_train, Y_train)
```

```python
w = linear_regression.coef_
b = linear_regression.intercept_
w,b
```

```python
y_pred = linear_regression.predict(X_valid.head())
y_pred
```

```python
pd.DataFrame({'y': Y_valid.head(), 'y_pred': y_pred})
```

```python
y_pred_total = linear_regression.predict(X_valid)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rmse = np.sqrt(mean_squared_error(Y_valid, y_pred_total))
r2_score = r2_score(Y_valid, y_pred_total)

print(f"RMSE: {rmse}")
print(f"R^2: {r2_score}")
```
